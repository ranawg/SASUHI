```python
import requests
from bs4 import BeautifulSoup

# Define the URL of the parent directory
url = "https://e4ftl01.cr.usgs.gov/MOLT/MOD11A1.061/2017.08.30/" ## need to adjust to download everthing 

# Send a GET request to fetch the webpage content
response = requests.get(url)
response.raise_for_status()  # Raise an error if the request fails

# Parse the HTML content
soup = BeautifulSoup(response.text, "html.parser")

# Extract all the links (anchor tags)
links = [link.get('href') for link in soup.find_all('a')]

# Filter out None and keep full URLs
full_links = [url + link for link in links if link is not None]

# Print the full links
print("Links found on the page:")
for full_link in full_links:
    print(full_link)
```
